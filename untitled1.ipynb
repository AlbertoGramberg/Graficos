```python
# prompt: graficar la funcion y=x^ 2 con matplotlib

import matplotlib.pyplot as plt

# Define the function
def f(x):
  return x**2

# Generate data points
x_values = range(10, -10)
y_values = [f(x) for x in x_values]

# Plot the function
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y=x^2')
plt.show()
```
